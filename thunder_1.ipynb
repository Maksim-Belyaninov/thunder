```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.load('/2012_thunderbolts_clastered.npz')
df = pd.DataFrame(data['strikes'])
for j in range(df.clnb.max()+1):
  Data_lon = []
  Data_lat = []
  Data_amp = []
  for i in range(len(df)):
    if df.at[i, 'clnb'] == j:
      Data_lon.append(df.at[i, 'lon'])
      Data_lat.append(df.at[i, 'lat'])
      Data_amp.append(np.abs(df.at[i, 'amp']))
  if len(Data_lon) > 0:
    plt.scatter(Data_lon, Data_lat, c=Data_amp, vmin=0, vmax=df.amp.max())
    plt.title('clnb '+str(j))
    plt.xlabel('lon'); plt.ylabel('lat')
    plt.grid()
    cb = plt.colorbar()
    cb.set_label('amp')
    plt.show()
```
